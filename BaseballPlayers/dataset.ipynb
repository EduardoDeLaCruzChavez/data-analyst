# Splitting the Field
## Objective
Your dataset contains the names of all the baseball players from the 2025 MLB season, along with their team(s) and position(s). A player qualifies for a position by starting at least 20 games at that position in the field or 5 games as a pitcher.

Your task is to aggregate this data and return a table counting the number of players at each position, sorted from highest to lowest.

[[Source]](https://mavenanalytics.io/data-drills/splitting-the-field)

```python
import pandas as pd

file = pd.read_csv("baseball_positions.csv")

file.head()
```

```python
positions = file["Position"].str.get_dummies("/")
positions.head(5)
```

```python
summ = positions.sum().sort_values(ascending=False)
result = summ.reset_index(name="Count")
result.columns = ["Position", "Total"]

print(f"third most popular position: {result["Total"][2]}")

result
```
